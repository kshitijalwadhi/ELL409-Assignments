```python
N_FOLDS = 10
```

```python
def cross_validation_split(dataset):
    dataset_split = list()
    dataset_copy = list(dataset)
    fold_size = int(len(dataset) / N_FOLDS)
    for i in range(N_FOLDS):
        fold = list()
        while len(fold) < fold_size:
            index = random.randrange(len(dataset_copy))
            fold.append(dataset_copy.pop(index))
        dataset_split.append(fold)
    return dataset_split
```

```python
def getDataset():
    dataset = []
    for i,row in df.iterrows():
        xi = []
        for j in range(0,26):
            xi.append(row[j])
        dataset.append(xi)
    return dataset
```

```python
dataset = getDataset()
```

```python
dataset_split = cross_validation_split(dataset)
```

```python
def evaluate(dataset_split,KERNEL,C):
    acc_list = []
    for fold in dataset_split:
        trainDS = list(dataset_split)
        trainDS.remove(fold)
        trainDS = sum(trainDS,[])
        testDS = list()
        
        for row in fold:
            rowCopy = list(row)
            testDS.append(rowCopy)
        
        trainX = [train[:-1] for train in trainDS]
        trainY = [train[-1] for train in trainDS]
        
        testX = [test[:-1] for test in testDS]
        testY = [test[-1] for test in testDS]
        
        prob = svm_problem(trainY,trainX)
        param = svm_parameter()
        param.kernel_type = KERNEL
        param.C = C
        m = svm_train(prob,param)
        p_label, p_acc, p_val = svm_predict(testY, testX, m)
        acc_list.append(p_acc[0])
    return np.mean(acc_list)
```

```python
evaluate(dataset_split,POLY,4)
```

```python
res = []
for KERNEL in [LINEAR,POLY,RBF,SIGMOID]:
    for C in np.arange(1,20,1):
        err = evaluate(dataset_split,KERNEL,C)
        res.append([KERNEL,C,err])
```
